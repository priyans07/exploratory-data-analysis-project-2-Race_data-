```python
# reset index
df2.reset_index(drop=True)
```

```python
#  fix types
df2.dtypes
```

```python
df2['athlete_age']=df2['athlete_age'].astype(int)
df2['Athlete average speed']=df2['Athlete average speed'].astype(float)
df2['Athlete ID']=df2['Athlete ID'].astype(int)
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
# rename columns
# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers    float64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed         object
# Athlete ID                   float64
# athlete_age                  float64
df2=df2.rename(columns={
    'Year of event':'year','Event dates':'race_day','Event name':'race_name','Event distance/length':'race_length','Event number of finishers':'race_number_of_finishers'
    ,'Athlete performance':'athlete_performance','Athlete gender':'athlete_gender','Athlete average speed':'athlete_average_speed',
    'Athlete ID':'athlete_id'

})
```

what age groups are best in the 50mi race (20+ race min)

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>19').head(15)
```

what age groups are worst in 50mi race (20+min race)

```python
df3.query('race_length=="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).query('count>19').sort_values('mean').head(10)
```

insight -: age group between 30-40 are mostly good in performing on racwes as compared to above 45 age group

```python
# seasons for the data -> slowerr in summer than winter
# spring 3-5
# fall-9-11
# winter-12-2
# split between 2 decimals
df3['race_month']=df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(10)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
# df3.query('race_season == "Spring"').head(10)
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count'])
```
